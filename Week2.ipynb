```python
# Creating 2D Geometry image using OpenCv. Start with creating Black image

black_img = np.zeros(shape=(512, 512, 3), dtype=np.int16)
plt.imshow(black_img)
```

```python
# Adding Rectangle defining Coordinate pt1 and pt2
cv.rectangle(black_img, pt1=(384, 0), pt2=(510, 150), color=(0, 255, 0), thickness=10)
plt.imshow(black_img)
```

```python
# Next, add a square shape according to the pt2 and pt2 coordinates
cv.rectangle(black_img, pt1=(200, 200), pt2=(300, 300), color=(0, 0, 255), thickness=15)
plt.imshow(black_img)
```

```python
#Next is to add a circle shape according to the radius!
cv.circle(black_img, center=(100, 100), radius=50, color=(255, 0, 0), thickness=8)
plt.imshow(black_img)
```

```python
# Adding diagonal line according to the coordinates of pt1 and pt2
cv.line(black_img, pt1=(0,0), pt2=(512, 512), color=(255, 255, 255), thickness=5)
plt.imshow(black_img)
```

```python
# Adding text with a font written with a predetermined size
font = cv.FONT_HERSHEY_SIMPLEX
cv.putText(black_img, text='Hello', org=(10, 500), fontFace=font, fontScale=4, color=(255, 255, 0), thickness=2, lineType=cv.LINE_AA)
plt.imshow(black_img)
```

```python
# Creating Black Image Again with int32 data type
black_img2 = np.zeros(shape=(512,512, 3), dtype=np.int32)
plt.imshow(black_img2)
```

```python
# Initialize a NumPy Array with the data type int32
vectices = np.array([[100, 300], [200, 200], [400, 300], [200, 400]], dtype=np.int32)
vectices
```

```python
# Reshape the array
pts = vectices.reshape((-1, 1, 2))
pts
```

```python
# Add a polyline to the seceond black image that has been created
cv.polylines(black_img2, [pts], isClosed=True, color=(255, 0, 0), thickness=5)
plt.imshow(black_img2)
```

7. Make rectanges and circles on the face of your photo image when you are active.

```python
# Load and Read new Image
img13 = cv.imread('/content/drive/MyDrive/PCVK/Images/chan.jpg')
img14 = cv.cvtColor(img13, cv.COLOR_BGR2RGB)

face_cascade = cv.CascadeClassifier(cv.data.haarcascades + 'haarcascade_frontalface_default.xml')
faces = face_cascade.detectMultiScale(img14, scaleFactor=1.1, minNeighbors=5)

for (x, y, w, h) in faces:
    # draw a rectangle around the face
    cv.rectangle(img14, (x, y), (x+w, y+h), (0, 0, 255), 5)
    # draw a circle at the center of the face
    cx, cy = x + w // 2, y + h // 2
    radius = min(w, h) // 2
    cv.circle(img14, (cx, cy), radius, (255, 0, 0), 5)

plt.imshow(img14)
```
