```python
#Import libraries
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt 
```

Dilation without OpenCV morphology library

```python
# Load the input image
image = cv.imread('/content/drive/MyDrive/PCVK/images/j.png', cv.IMREAD_GRAYSCALE)

# Define the structuring element
structuring_element = np.ones((5, 5))

# Define the output image
output_image = np.zeros_like(image)

# Slide the structuring element over the image and perform the dilation operation
for i in range(image.shape[0]):
    for j in range(image.shape[1]):
        if image[i, j] == 255:
            output_image[i:i+5, j:j+5] = np.logical_or(output_image[i:i+5, j:j+5], structuring_element)

# Visualize the input and output images
fig, axs = plt.subplots(1, 2, figsize=(8, 4))
axs[0].imshow(image, cmap='gray')
axs[0].set_title('Input Image')
axs[1].imshow(output_image, cmap='gray')
axs[1].set_title('Dilated Image')
plt.show()
```

Erosion Operation without OpenCV morphology library

```python
# Load the input image
image = cv.imread('/content/drive/MyDrive/PCVK/images/j.png', cv.IMREAD_GRAYSCALE)

# Define the structuring element
structuring_element = np.ones((5, 5))

# Define the output image
output_image = np.zeros_like(image)

# Slide the structuring element over the image and perform the erosion operation
for i in range(2, image.shape[0]-2):
    for j in range(2, image.shape[1]-2):
        if (image[i-2:i+3, j-2:j+3] * structuring_element).min() == 255:
            output_image[i,j] = 255

# Visualize the input and output images
fig, axs = plt.subplots(1, 2, figsize=(8, 4))
axs[0].imshow(image, cmap='gray')
axs[0].set_title('Input Image')
axs[1].imshow(output_image, cmap='gray')
axs[1].set_title('Eroded Image')
plt.show()
```

Dilation Operation with OpenCV morphology library

```python
# Load the input image
image = cv.imread('/content/drive/MyDrive/PCVK/images/j.png', cv.IMREAD_GRAYSCALE)

# Define the structuring element
structuring_element = cv.getStructuringElement(cv.MORPH_RECT, (5, 5))

# Perform the dilation operation using the structuring element
output_image = cv.dilate(image, structuring_element)

# Visualize the input and output images
fig, axs = plt.subplots(1, 2, figsize=(8, 4))
axs[0].imshow(image, cmap='gray')
axs[0].set_title('Input Image')
axs[1].imshow(output_image, cmap='gray')
axs[1].set_title('Dilated Image')
plt.show()
```

Erosion with OpenCV morphology library

```python
# Load the input image
image = cv.imread('/content/drive/MyDrive/PCVK/images/j.png', cv.IMREAD_GRAYSCALE)

# Define the structuring element
structuring_element = np.ones((5, 5))

# Perform erosion operation using OpenCV morphology library
output_image = cv.erode(image, structuring_element)

# Visualize the input and output images
fig, axs = plt.subplots(1, 2, figsize=(8, 4))
axs[0].imshow(image, cmap='gray')
axs[0].set_title('Input Image')
axs[1].imshow(output_image, cmap='gray')
axs[1].set_title('Eroded Image')
plt.show()
```

Opening operation using both OpenCV morphology library and without

a) Using OpenCV morphology library

```python
# Load the input image
image = cv.imread('/content/drive/MyDrive/PCVK/images/j.png', cv.IMREAD_GRAYSCALE)

# Define the structuring element
structuring_element = cv.getStructuringElement(cv.MORPH_RECT, (7, 7))

# Perform the opening operation using morphologyEx function
opening_image = cv.morphologyEx(image, cv.MORPH_OPEN, structuring_element)

# Visualize the input and output images
fig, axs = plt.subplots(1, 2, figsize=(8, 4))
axs[0].imshow(image, cmap='gray')
axs[0].set_title('Input Image')
axs[1].imshow(opening_image, cmap='gray')
axs[1].set_title('Opened Image')
plt.show()
```

b) Without using OpenCV morphology library

```python
# Load the input image
image = cv.imread('/content/drive/MyDrive/PCVK/images/j.png', cv.IMREAD_GRAYSCALE)

# Define the structuring element
structuring_element = np.ones((7, 7))

# Perform the erosion operation
erosion_image = np.zeros_like(image)
for i in range(image.shape[0]):
    for j in range(image.shape[1]):
        if image[i, j] == 255:
            erosion_image[i:i+7, j:j+7] = np.logical_or(erosion_image[i:i+7, j:j+7], structuring_element)

# Perform the dilation operation
dilation_image = np.zeros_like(image)
for i in range(erosion_image.shape[0]):
    for j in range(erosion_image.shape[1]):
        if erosion_image[i, j] == 1:
            dilation_image[i:i+7, j:j+7] = 1

# Visualize the input and output images
fig, axs = plt.subplots(1, 2, figsize=(8, 4))
axs[0].imshow(image, cmap='gray')
axs[0].set_title('Input Image')
axs[1].imshow(dilation_image, cmap='gray')
axs[1].set_title('Opened Image')
plt.show()
```

Closing operation using both OpenCV morphology library and without

a) Using OpenCV Morphology Library

```python
# Load the input image
image = cv.imread('/content/drive/MyDrive/PCVK/images/j.png', cv.IMREAD_GRAYSCALE)

# Define the structuring element
structuring_element = cv.getStructuringElement(cv.MORPH_RECT, (7, 7))

# Perform the closing operation with OpenCV
closing_image = cv.morphologyEx(image, cv.MORPH_CLOSE, structuring_element)

# Visualize the input and output images
fig, axs = plt.subplots(1, 2, figsize=(8, 4))
axs[0].imshow(image, cmap='gray')
axs[0].set_title('Input Image')
axs[1].imshow(closing_image, cmap='gray')
axs[1].set_title('Closed Image (OpenCV)')
plt.show()
```

Dilation operation and show the results on the image of the Structuring
Element shaped 3x3 and 5x5 Cross using the OpenCV morphology library.

```python
# Load the input image
image = cv.imread('/content/drive/MyDrive/PCVK/images/j.png', cv.IMREAD_GRAYSCALE)

# Define the structuring element (3x3 cross)
structuring_element_3x3 = cv.getStructuringElement(cv.MORPH_CROSS, (3,3))

# Define the structuring element (5x5 cross)
structuring_element_5x5 = cv.getStructuringElement(cv.MORPH_CROSS, (5,5))

# Perform the dilation operation with a 3x3 cross-shaped structuring element
dilation_image_3x3 = cv.dilate(image, structuring_element_3x3, iterations=1)

# Perform the dilation operation with a 5x5 cross-shaped structuring element
dilation_image_5x5 = cv.dilate(image, structuring_element_5x5, iterations=1)

# Visualize the input and output images
fig, axs = plt.subplots(2, 2, figsize=(10, 8))
axs[0, 0].imshow(image, cmap='gray')
axs[0, 0].set_title('Input Image')
axs[0, 1].imshow(dilation_image_3x3, cmap='gray')
axs[0, 1].set_title('Dilated Image (3x3 cross)')
axs[1, 0].imshow(image, cmap='gray')
axs[1, 0].set_title('Input Image')
axs[1, 1].imshow(dilation_image_5x5, cmap='gray')
axs[1, 1].set_title('Dilated Image (5x5 cross)')
plt.show()
```
